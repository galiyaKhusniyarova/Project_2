```python
import pandas as pd
import numpy as np
import os
import glob
from pathlib import Path
import datetime as dt
%matplotlib inline
```

## Prepare the data

```python
realtor_data_df = pd.read_csv(
    Path("./realtor-data.csv"),
    index_col = 'sold_date',
    parse_dates = True,
    infer_datetime_format = True
)
```

```python
display(realtor_data_df)
```

```python
realtor_data_df = realtor_data_df[realtor_data_df.index.notnull()]
display(realtor_data_df)
```

## Slice the data

```python
realtor_data_2015 = realtor_data_df.loc['2015'].sort_index()
display(realtor_data_2015)
```

```python
realtor_data_2016 = realtor_data_df.loc['2016'].sort_index()
display(realtor_data_2016)
```

```python
realtor_data_2017 = realtor_data_df.loc['2017'].sort_index()
display(realtor_data_2017)
```

```python
realtor_data_2018 = realtor_data_df.loc['2018'].sort_index()
display(realtor_data_2018)
```

```python
realtor_data_2019 = realtor_data_df.loc['2019'].sort_index()
display(realtor_data_2019)
```

## Download to csv file

```python
# Year 2018
realtor_data_2018.to_csv(Path('./realtor_data/2018_realtor_data.csv'))

# Year 2019
realtor_data_2019.to_csv(Path('./realtor_data/2019_realtor_data.csv'))
```
